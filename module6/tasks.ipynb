## Task: GamePlatofrm Preference 

You are interested in creating your own video game. Analyze a dataset from IGN which ranks the different platforms and game types

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

## Load and Review Data

For this anlysis we will use the : `data/ign_scores.csv`
- Use the `Platofrm` column as index

```python
# Load the data in to a DataFrame
file_path = '../data/ign_scores.csv'
ign_data = pd.read_csv(file_path, index_col='Platform')

# Print the data
ign_data.head()
```

### Task: Which Platform is Best? 

Create a `bar chart` that shows the average score for `racing` games, or each platform.

```python
# TODO: Create a bar chart that shows the average score for racing games, grouped by platform.
plt.figure(figsize=(10, 6))
sns.barplot(x=ign_data.index, y=ign_data['Racing'])
plt.xlabel('Platform')
plt.xticks(rotation=90)
plt.ylabel('Average Score')
plt.title('Average Score for Racing Games, Grouped by Platform')
plt.show()
```

### Task: WHich one is the Best One for Racing Games

If `Wii` platform is not the best, which platform is?

```python
# TODO: Get the best platform for racing games
# Use the idxmax() method to get the index of the row with the highest value in the 'Racing' column
best_platform = ign_data['Racing'].idxmax()
print(f'The best platform for racing games is \033[32m{best_platform}\033[0m')
print(f'with an average score of \033[35m{ign_data.loc[best_platform, "Racing"]:.2f}\033[0m')
```

### Task: All possible combinations 
Create a Heatmap of the average scores by genre in each cell

```python
# TODO: Create a heatmap that shows the average score for each genre on each platform.
plt.figure(figsize=(14, 6))
sns.heatmap(ign_data, annot=True, fmt='.1f', cmap='coolwarm')
plt.xlabel('Genre')
plt.ylabel('Platform')
plt.title('Average Score for Each Genre on Each Platform')
plt.show()
```

```python
# TODO: Get the best platform for each genre
# Use the idxmax() method to get the index of the row with the highest value in each column
best_platforms = ign_data.idxmax()
print('The \033[32mbest\033[0m platform for each genre is:')
for genre, platform in best_platforms.items():
    print(f'\033[36m{genre}\033[0m\n \033[35m{platform}\033[0m: \033[32m{ign_data.loc[platform, genre]:.2f}\033[0m\n')
```

```python
#TODO: Get the worst platform for each genre
worst_platforms = ign_data.idxmin()
print('The \033[31mworst\033[0m platform for each genre is:')
for genre, platform in worst_platforms.items():
    print(f'\033[36m{genre}\033[0m\n \033[35m{platform}\033[0m: \033[31m{ign_data.loc[platform, genre]:.2f}\033[0m\n')
```

```python
# TODO: Get the highest comination of genre and platform and lowest combination of genre and platform
# Use the stack() method to convert the DataFrame to a Series with a MultiIndex
# Use the idxmax() and idxmin() methods to get the index of the row with the highest and lowest value in the Series
best_combination = ign_data.stack().idxmax()
worst_combination = ign_data.stack().idxmin()
print(f'The \033[32mbest\033[0m combination of genre and platform is:')
# Use the loc[] method to get the value of the row with the highest and lowest value in the Series
print(f'\033[36m{best_combination[1]}\033[0m on \033[35m{best_combination[0]}\033[0m with a score of \033[32m{ign_data.loc[best_combination[0], best_combination[1]]:.2f}\033[0m\n')
print(f'The \033[31mworst\033[0m combination of genre and platform is:')
print(f'\033[36m{worst_combination[1]}\033[0m on \033[35m{worst_combination[0]}\033[0m with a score of \033[31m{ign_data.loc[worst_combination[0], worst_combination[1]]:.2f}\033[0m\n')
```

```python
# TODO: Create a scatter pelot that shows the correlation between `sugarpercent` and `winpercent`.
plt.figure(figsize=(10, 6))
sns.scatterplot(x=candy_data['sugarpercent'], y=candy_data['winpercent'])
plt.xlabel('Sugar Percentage')
plt.ylabel('Win Percentage')
plt.title('Correlation Between Sugar Percentage and Win Percentage')
plt.show()
```

```python
# TODO: Create a regplot to show the correlation between `sugarpercent` and `winpercent`.
plt.figure(figsize=(10, 6))
sns.regplot(x=candy_data['sugarpercent'], y=candy_data['winpercent'])
plt.xlabel('Sugar Percentage')
plt.ylabel('Win Percentage')
plt.title('Correlation Between Sugar Percentage and Win Percentage')
plt.show()

```
