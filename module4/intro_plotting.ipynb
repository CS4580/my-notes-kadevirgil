## Bar Plot

## Area Plot
Useful for when you want to understand how values develop and are composed over time

```python
# Load data set 
enrollment_df = pd.read_csv('../data/enrollment.csv')
enrollment_df.head()
```

```python
# Create an area plot of the enrollments
enrollment_df.plot(kind='area', x='Year', title='Enrollment in Schools', figsize=(10, 6))
```

```python
# Data is not good because it is not organized by year
# We want to see the total enrollment in each year and how it changes over time
# We will sort the data by year, then plot it using pivot method in pandas
pivot_df = enrollment_df.pivot(index='Year', columns='Class', values='Enrollment')
pivot_df.head()

# Plot the pivot data
pivot_df.plot(kind='area', title='Enrollment in Schools by Class (Stacked)', figsize=(10, 6))
```

```python
# The data is stacked by default
# If you want to unstack the data, you can set stacked=False
pivot_df.plot(kind='area', title='Enrollment in Schools by Class (Unstacked)', figsize=(10, 6), stacked=False)
```

## Correlation Analysis
- What is correlation Analysis
- How does correlation analysis help with data cleaning?
- Coding Example

### Correlation Analysis
is a statistical technique used to examine the strength and direction of the relationship
    between two or more variables

Analyze the degree to which changes from one variable are associated with changes from antoher variable

### How to do it? 
Use `Correlation Coefficients`, which is a measurement of the strength and the direction of the relationship between the two variables

$$
\{Y, X\}
$$

### Popular Correlation Coefficients

#### Pearson
- Used for continuous data 
- Measures the strength of the `linear relationship` between the variables
- Sensitive to outliers

#### Spearman
- Use for ordinal or ranked data
- Measures the strength of the `monotonic relationhship` between the variables, which can be linear or non-linear
- More robust towards outliers

Correlation analysis can identify variables that are highly correlated to each other.

The analysis may indicate if one variable is `redundant` and can be eliminated

### Dealing with Categorical Data
- data has many `non-numeric` features. You CAN NOT feed them to a learning model. They need to be converted.
- Use the  `pd.dtypes()` to see the data types 

### Two main types 
- Label encoding 
- One-hot encoding

### Label encoding
Each `unique` category in the categorical variable is assigned a numerical label. Typically starting at 0, 1, 2 .... etc

### One-hot encoding
A new binary feature is created `for each` category in the vategorical variable. The value of that feature is a set to 1 if the observation belongs to that. 

```python
# See dataset dtypes
data_df = pd.read_csv('../data/hotel_bookings.csv')

# Print the data types of each column
data_df.dtypes
```

```python
# TODO: Create a list of those columns that are of type 'object'
data_df.select_dtypes(include=['object']).head()
# List comprehension approach
# obj_cols = [col for col in data_df.columns if data_df[col].dtype == 'object']
```

```python
# TODO: Create a new dataframe from only those columns 
# that are of type 'object'
data_obj_df = data_df.select_dtypes(include=['object'])
data_obj_df.head()
```

```python
# TODO: Filter the data now to filter the columns that are of type 'object'
# using the dataframe created in the previous step
cat_df = data_df[data_obj_df.columns]
cat_df.head()
```

```python
# TODO: Print the unqiue values those columns that are of type 'object'
for col in cat_df.columns:
    print(f"Column: {col}\n{cat_df[col].unique()}\n")
```

```python
# TODO: Cast the categorical features to the vategory date type ofr 'arrival_date_month'
# January -> 1, February -> 2, etc.
month_map = {
    'January': 1,
    'February': 2,
    'March': 3,
    'April': 4,
    'May': 5,
    'June': 6,
    'July': 7,
    'August': 8,
    'September': 9,
    'October': 10,
    'November': 11,
    'December': 12
}

# One/Old approach to map values
# cat_df['arrival_date_month'] = cat_df['arrival_date_month'].map(month_map)

# New approach to map values using .loc
# cast it to caegory data type
cat_df.loc[:, 'arrival_date_month'] = cat_df['arrival_date_month'].map(month_map)

# Print the unique values of the column to verify the mapping
cat_df.head()
```

```python
# TODO: Now do the country hotel feature using label encoding from sklearn
from sklearn.preprocessing import LabelEncoder

# Create a label encoder object to use for encoding the country feature
le = LabelEncoder()

# Fit the label encoder object on the country feature using fit_transform()
# cat_df['country'] = le.fit_transform(cat_df['country']) # this gives a warning because of the copy of the data
cat_df.loc[:, 'country'] = le.fit_transform(cat_df['country'])
```
