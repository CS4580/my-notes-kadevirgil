```python
import pandas as pd 
# Load data
reviews_df = pd.read_csv('../data/winemag-data-130k-v2.csv')
reviews_df.head()
```

```python
# TODO: Who are the most common wine reviewers
# Create a series whose index is the 'taster_twitter_handle' column
# WHose values count now many reviews each person wrote
reviews_df['taster_twitter_handle'].value_counts()

# You can also do this with groupby()
# reviews_df.groupby('taster_twitter_handle').size()
# OR the count() method
# reviews_df.groupby('taster_twitter_handle').taster_twitter_handle.count()
```

```python
# TODO: What is the best wine I can bhy for a given amount of money?
# Create a Serires whose index is the wine price and whose values are the max number of points
# a wine costing htat much was given in a review
# Sort the series by price, ascending ($4 at the top)
best_wine_by_price = reviews_df.loc[reviews_df.groupby('price')['points'].idxmax()]
best_wine_by_price = best_wine_by_price[['price', 'title', 'points']].sort_values('price')
best_wine_by_price.reset_index(drop=True, inplace=True)
best_wine_by_price
```

```python
# TODO: What are the minimum and maximum prices for each variety of wine?
# Create a DataFrame whose index is the variety and whose values are min and max values
price_extremes = reviews_df.groupby('variety')['price'].agg(['min', 'max'])
price_extremes 
```

```python
# TODO: What are the most expesive wine varieties?
# Create a variable sorted_varieties containing a copy of the dataframe from the previous question
# where varieties are sorted in descending order based on minimum price, then on maximum price
sorted_varieties = price_extremes.sort_values(by=['min', 'max'], ascending=False)
sorted_varieties
```

```python
# TODO: Create a Series whose index is reviewers and whose values is the average 
# review score given out by that reviewer
# Sort the values in the Series in descending order to get the reviewer with the highest average rating
reviewer_mean_ratings = reviews_df.groupby('taster_name')['points'].mean().sort_values(ascending=False)
reviewer_mean_ratings
```
